# 测试一下speech2text

```python
import base64
import utils_audio
from IPython.display import Audio

data = {"sample_rate":16000,
 "channels":1,
 "language":"zh",
 "audio":"",
 "ts":1694528463}

len(data['audio'])
sr = data['sample_rate']
audio_buffer=base64.b64decode(data['audio'])
utils_audio.save_audio_buffer(audio_buffer,sr, fp="tmp.wav")
Audio("tmp.wav")
```

```python
import torch
from speech2text import Speech2Text

DEVICE = "cuda:0" if torch.cuda.is_available() else "cpu"
M_s2t = Speech2Text(model_type="tiny", download_root="/root/autodl-fs/whisper", device=DEVICE)
M_s2t.init()
text = M_s2t.transcribe_buffer(base64.b64decode(data['audio']),
                               sr_inp=data['sample_rate'],
                               channels_inp=data['channels'],
                               language=data['language'],
                               fp16=False)
text
```

**合成然后再识别**

```python
from IPython.display import Audio
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"  # 多行都显示
import utils_audio
from text2speech import Text2Speech as Text2Speech_vits
import torch
from speech2text import Speech2Text

M_tts = Text2Speech_vits(model_dir="/root/autodl-fs/VITS_DATA/OUTPUT_MODEL_四郎_daniel_bruce/G_latest.pth",
                         config_fp="/root/autodl-fs/VITS_DATA/OUTPUT_MODEL_四郎_daniel_bruce/config.json").init()

sr, wav = M_tts.tts_fn(text="测试一下",
                         speaker="bruce",
                         language="auto",  # 用Mix的话就相当于直接读字母发音了
                         text_cleaners=M_tts.hparams['data']['text_cleaners'],
                         speed=1)
utils_audio.save_audio(wav, sr, fp="./tmp.wav")
Audio("./tmp.wav")

audio_buffer=wav.tobytes()
audio2req = base64.b64encode(audio_buffer).decode()
len(audio2req)


DEVICE = "cuda:0" if torch.cuda.is_available() else "cpu"
M_s2t = Speech2Text(model_type="tiny", download_root="/root/autodl-fs/whisper", device=DEVICE).init()
text = M_s2t.transcribe_buffer(base64.b64decode(audio2req),
                               sr_inp=sr,
                               channels_inp=1,
                               language="zh",
                               fp16=False)
print(text)
```
