# 各speaker声音测试

```python
for speaker in list(M_tts.hparams['speakers'].keys())[:10]:
    print(">> speaker: %s" % speaker)
    sr, wav = M_tts.tts_fn(text="致以诚挚的问候和美好的祝愿",
                             speaker=speaker,
                             language="auto",  # 用Mix的话就相当于直接读字母发音了
                             speed=1)
    utils_audio.save_audio(wav, sr, fp="./mock_vits_%s.wav" % speaker)
    Audio("./mock_vits_%s.wav" % speaker)
```
