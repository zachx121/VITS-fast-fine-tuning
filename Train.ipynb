检查一下各个角色的音频数据长度

````python
import librosa
import os

for speaker in os.listdir("./segmented_character_voice"):
    duration=0
    cnt=0
    for filename in os.listdir(f"./segmented_character_voice/{speaker}"):
        duration += librosa.get_duration(filename=f"./segmented_character_voice/{speaker}/{filename}")
        cnt+=1
    print("[speaker] %s [sample_count] %s [duration] %d" % (speaker, cnt, duration))
````
