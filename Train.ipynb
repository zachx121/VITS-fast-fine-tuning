# 推理
- 直接使用`OUTPUT_MODEL`里的`G_latest.pth`和`config.json`就行
- 配合serving分支的text2speech.py

````python
from IPython.display import Audio
import utils_audio
from text2speech import Text2Speech as Text2Speech_vits

M_tts = Text2Speech_vits(model_dir="/root/VITS-fast-fine-tuning/OUTPUT_MODEL/G_latest.pth",
                    config_fp="/root/VITS-fast-fine-tuning/OUTPUT_MODEL/config.json").init()

````

````python
sr, wav = M_tts.tts_fn(text="这里是一段测试音频1234，我是桑达科技的创始人程晓敏，吹风磁暴救救我 abcd. good time.",
                         speaker="四郎配音",
                         language="auto",  # 用Mix的话就相当于直接读字母发音了
                         text_cleaners=["chinese_cleaners"],
                         speed=1.0)

utils_audio.save_audio(wav, sr, fp="./mock_vits.wav")
Audio("./mock_vits.wav")
````
